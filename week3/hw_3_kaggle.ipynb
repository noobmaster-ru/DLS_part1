```python
# где TotalSpent = " " заменим нулями, так как они ещё не потратили денег -
#     у них у всех ClientPeriod == 0
def replace_gap_to_zero(data: pd.DataFrame) -> pd.DataFrame:
  data['TotalSpent'] = data['TotalSpent'].replace(' ', np.nan)
  data['TotalSpent'] = pd.to_numeric(data['TotalSpent'], errors='coerce')

  # заполняем пробелы нулями
  data['TotalSpent'] = data['TotalSpent'].fillna(0)
  data["TotalSpent"] = data["TotalSpent"].astype(float)
  return data
```

## One-hot-encoding

```python
# YOUR CODE
cat_cols_clean = [item for item in cat_cols if item not in binary_cols]
print(cat_cols_clean)

# делаем one hot encoding категориальных признаков
categorical_data = data[cat_cols_clean]
ohe_encoder = OneHotEncoder(sparse_output=False)
encoded_data = ohe_encoder.fit_transform(categorical_data)
encoded_df = pd.DataFrame(encoded_data, columns=ohe_encoder.get_feature_names_out(cat_cols_clean))

# print(df_main.head(5),"\n")
data_ohe = data.drop(columns=cat_cols_clean, axis=1)

data_ohe = pd.concat([data_ohe,encoded_df], axis=1)
```

### Делаем нормализацию и делим выборку на test/val

```python
# делим выборку на тренировочную и тестовую
y = data_ohe[target_col]
X = data_ohe.drop(columns=[target_col])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.8,
    random_state = 42
)

# делаем нормализацию
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Обучаем модели

```python
# обучаем модели
logr_cv = LogisticRegressionCV(
    cv=5,
    random_state=42,
    Cs=[100, 20, 15, 10, 5,  1, 0.1, 0.01, 0.001],
    scoring="roc_auc",
    refit=True,
    class_weight='balanced'
)

logr_cv.fit(X_train_scaled, y_train)


y_pred = logr_cv.predict_proba(X_test_scaled)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred)
print(f"test roc_auc_score = {roc_auc:.4f}")
print("best_C = ",logr_cv.C_)
```
